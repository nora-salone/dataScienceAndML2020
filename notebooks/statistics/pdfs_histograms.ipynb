```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def binWidth(bins,xmin,xmax):
    return (xmax-xmin)/bins
```

```python
def gauss(x,mean,sigma):
    return 1./math.sqrt(2 *math.pi * sigma**2) * math.exp(-(x-mean)**2/(2*sigma**2))
```

```python
mu = 0
sigma = 3
xmin = -15
xmax = 15
```

```python
N=100
bins = 30
sample = np.random.normal(loc=mu,scale=sigma,size=N)
plt.hist(sample, bins=bins, range=[xmin,xmax])
plt.gca().set_title('number of bins: %i, bin width:%f'%(bins,binWidth(bins,xmin,xmax)))
```

```python
#generate gaussian pdf
xVals = np.linspace(xmin,xmax,1000)
yVals = [gauss(x,mu,sigma) for x in x]
```

```python
N=10000000
bins = 1000
sample = np.random.normal(loc=mu,scale=sigma,size=N)
plt.hist(sample, bins=bins, range=[xmin,xmax], density=True)
plt.gca().set_title('number of bins: %i, bin width:%f'%(bins,binWidth(bins,xmin,xmax)))
plt.plot(xVals,yVals)
plt.show()
```
